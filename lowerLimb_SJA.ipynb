##Vamos calcular os ângulos do membro inferior no plano sagital

```python
import numpy as np
from IPython.display import display, Latex
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = 'data/WinterTableA1.txt'
data = np.loadtxt(filename, skiprows=2, unpack=False)
markers = ['RIB CAGE', 'HIP', 'KNEE', 'FIBULA', 'ANKLE', 'HEEL', 'MT5', 'TOE']
```

```python
np.size(data,1)
```

```python
## shift de 2 colunas e calculo das diferenças das matrizes
segsCoord = data[:, 2:16] - data[:, 4:18]
```

###Primeiro os Angulos dos segmentos

```python
#coxa
tetha21 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,3],segsCoord[:,2])))

tetha32 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,5],segsCoord[:,4])))

#perna
tetha43 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,7],segsCoord[:,6])))

tetha54 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,9],segsCoord[:,8])))

#meioPe
tetha65 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,11],segsCoord[:,10])))

#antePe
tetha76 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,13],segsCoord[:,12])))
```

```python
plt.clf()
plt.figure(figsize=(9,12))
ax1 = plt.subplot(4,1,1)
plt.title('Angulos dos Segmentos')
plt.plot(data[:,1], tetha21)
plt.ylabel('Coxa')
plt.grid()

plt.subplot(4,1,2, sharex=ax1)
plt.plot(data[:,1], tetha43)
plt.ylabel('Perna')
plt.grid()

plt.subplot(4,1,3, sharex=ax1)
plt.plot(data[:,1], tetha65)
plt.ylabel('MeioPe')
plt.grid()

plt.subplot(4,1,4, sharex=ax1)
plt.plot(data[:,1], tetha76)
plt.ylabel('Antepe')
plt.xlabel('tempo(s)')
plt.grid()
```

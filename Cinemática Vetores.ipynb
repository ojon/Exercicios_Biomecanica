Alguns cálculos simples com vetores, temos um corpo que se move de $\vec{A}$ para $\vec{B}$ para $\vec{C}$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#vetores posição
A = np.array([1, 2, 3])
B = np.array([3, 2, 1])
C = np.array([0, 0, 0])

# diferença dos tempos
deltaT = np.array([2, 1])
```

```python
# tempo (instântes)
t = np.append(0, deltaT.cumsum())
```

```python
distA_B = np.linalg.norm(B - A)
distB_C = np.linalg.norm(C - B)
distTotal = distA_B + distB_C

print distA_B 
print distB_C
print distTotal
```

    A distância de A a B é de 2.83m.

    A distância de B a C é de 3.74m.

    A distância total é de 6.57m.

```python
vA_B = distA_B/deltaT[0]
vB_C = distB_C/deltaT[1]
print vA_B
print vB_C
```

    A velocidade de A a B é de 1.41m/s
    A velocidade de B a C é de 3.74m/s

```python
aMean = (vB_C - vA_B)/deltaT[1]
print aMean
```

A aceleração média de $\vec{B}$ para $\vec{C}$ é de $2.33m/s^2$

```python
plt.rc('axes',  labelsize=14,  titlesize=14) 
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10) 
f, axarr = plt.subplots(3, 1, sharex = True, sharey=True, figsize=(14,7))
plt.suptitle('Cinematica Vetorial', fontsize=20);
axarr[0].plot(t, [A[0], B[0], C[0]], 'k', linewidth=4, label='$x(t)=1$')
axarr[1].plot(t, [A[1], B[1], C[1]], 'k', linewidth=4, label='$y(t)=0$')
axarr[2].plot(t, [A[2], B[2], C[2]], 'k', linewidth=4, label='$z(t)=0$')
axarr[0].set_ylabel('x(t) [m]')
axarr[1].set_ylabel('y(t) [m]')
#axarr[1].set_ylim(14)
axarr[2].set_ylabel('z(t) [m]')
#axarr[2].set_ylim(20)
plt.xlabel('tempo(s)')
```

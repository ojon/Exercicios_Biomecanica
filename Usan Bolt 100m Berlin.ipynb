Vamos fazer alguns calculos de valores cinemáticos com os dados da corrida de Usan Bolt:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


## tempos largada + parciais a cada 10m
timesArray=np.array([.146, 1.89, 2.88, 3.78, 4.64, 5.47, 6.29, 7.1, 7.92, 8.75, 9.58])

#calculo dos intervalos de tempos
timeDiff = (timesArray[1:] - timesArray[:-1])

#velocidade e aceleração
velocity  = np.append(0, 10/timeDiff)
aceleration = np.append(0, (velocity[1:] - velocity[:-1])/timeDiff)


```

```python
df = pd.DataFrame({'velocidade(m/s)' : velocity, 
              'aceleração(m/s²)' : aceleration}, 
             index = timesArray)
df
```

```python
displacenment = np.array(range(11))*10

plt.rc('axes',  labelsize=14,  titlesize=14) 
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10) 
f, axarr = plt.subplots(3, 1, sharex = True, figsize=(14,7))
plt.suptitle('Cinematica Usan Bolt 100m -  2009 Berlin', fontsize=20);
axarr[0].plot(timesArray, displacenment, 'k', linewidth=4, label='$r(t)=1$')
axarr[1].plot(timesArray,  velocity, 'b', linewidth=4, label='$v(t)=0$')
axarr[2].plot(timesArray,  aceleration, 'r', linewidth=4, label='$a(t)=0$')
axarr[0].set_ylabel('r(t) [m]')
axarr[1].set_ylabel('v(t) [m/s]')
#axarr[1].set_ylim(14)
axarr[2].set_ylabel('a(t) [m/s$^2$]')
#axarr[2].set_ylim(20)
plt.xlabel('tempo(s)')
```

```python
plt.rc('axes',  labelsize=14,  titlesize=14) 
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10) 
f, axarr = plt.subplots(2, 1, sharex = True, figsize=(14,7))
plt.suptitle('Cinematica Usan Bolt 100m -  2009 Berlin', fontsize=20);
axarr[0].plot(displacenment,  velocity, 'b', linewidth=4, label='$v(t)=0$')
axarr[1].plot(displacenment,  aceleration, 'r', linewidth=4, label='$a(t)=0$')
axarr[0].set_ylabel('v(t) [m/s]')
#axarr[1].set_ylim(14)
axarr[1].set_ylabel('a(t) [m/s$^2$]')
#axarr[2].set_ylim(20)
plt.xlabel('distancia(m)')
```

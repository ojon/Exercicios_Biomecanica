## A2:S40 flexão cotovelo

```python
A2S40img = Image(filename='A2S40.jpeg')
```

```python
A2S40img
```

as váriaveis são: F módulo da força do musculo, J_x componente x da força da articulação, J_y componente y da força da articulação e W peso do membro superior

```python
F, J_x, J_y, W = sym.symbols("F J_x J_y W")
```

A soma dos momentos em relação ao ponto corresponde ao centro da articulação do ombro é igual a 0 e o peso do membro superior é 6% do peso corporal

```python
W = 6
eq = F*0.05 - W*0.38
```

Resoltendo a equação para F

```python
sym.solve(eq,F)
```

Assim  $F = 45,6 \% PC$

```python
F = 45.6
```

Não existe translação ao longo do eixo horizontal, logo a soma das projeções das forças ao longo do eixo horizontal é igual a 0.

```python
eq = -F * np.cos(np.pi/6) - J_x
eq
```

Assim a componente $J_x = −39.4907584125704 \% PC$

O valor negativo significa que esta componente difere do desenho da figura, apontando para o sentido oposto, i.e para do ombro p/ mão 

No eixo vertical tambpem não existe translação:

```python
eq = F * np.sin(np.pi/6) + J_y - W
eq
```

Assim a componente $J_y = −16.8 \% PC$

Concluimos que a força $\vec{F}$ possui módulo de $45,6\%PC$, direção e sentido de acordo com a figura e a força $\vec{J} = (39.4908\%PC, -16.8\%PC)$

obs: a convençao do sistema de coordenadas adotado considera compomentes em x positivas quando apontam da esquerda p/ direta e componente y positivas quando apontam de baixo para cima

### cotovelo flexionado $90^o$

Para o caso do cotovelo flexionado o peso do membro superior está dividido entre o peso dos segmentos braço e antebração/mão

```python
F, Wa, Wf = sym.symbols("F Wa Wf")
Wa = 3
Wf = 3.5
```

```python
eq = F*0.05 - Wa*0.15 - 0.4*Wf
```

```python
sym.solve(eq,F)
```

Assim  $F = 37\% PC$

```python
F = 37
```

Não existe translação ao longo do eixo horizontal, logo a soma das projeções das forças ao longo do eixo horizontal é igual a 0.

```python
eq = -F * np.cos(np.pi/6) - J_x
eq
```

Assim a componente $J_x = −32.0429399400242 \% PC$

No eixo vertical também não existe translação:

```python
eq = F * np.sin(np.pi/6) + J_y - Wa - Wf
eq
```

Assim a componente $J_y = −12,0\% PC$

No final temos:

$F = 37\%PC$ direção e sentido de acordo com a figura e $\vec{J} = (32.0429\%PC, -12\%PC)$
